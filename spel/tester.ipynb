### My idea is to learn following these steps
1. pick the n (~3) first avd from X_test_all into X_test and y_test
2. run catBoost
3. collect result-data
4. move 1 avd from test_all into X_train and y_train
5. start from 1

```python
def get_rows_with_dates(df,n):
    # get the first rows in df with the n first dates
    dates = df.datum.unique()[:n]
    rows = df[df.datum=='ingen datum']   # empty with the same columns as df
    for d in dates:
        rows = pd.concat([rows,df[df.datum==d]], copy=True, ignore_index=True)
    print('rows',rows.shape)
    return rows
```

```python
def move_rows_from_A_to_B(A,B,n):
    # select the n first datum in A
    move = get_rows_with_dates(A,n)
    a=A.iloc[move.shape[0]:] # remove the n first datum
    
    # add to the end of B
    b = pd.concat([B,move],ignore_index=True,copy=True)
    return a,b    
```

```python
def initsplit_my_data(df,mitt_datum):
    train=df[df.datum<=mitt_datum]
    test_all = df[df.datum>mitt_datum]

    print('Check that this is True',test_all.shape[0]+train.shape[0] == df.shape[0])
    train.shape,test_all.shape

    return train,test_all,
```

```python
# CatBoost fit
# save result
# move m (~1) rows from test_all to X_train
# Do it again date for date
```

```python
def catfit(trainPool, testPool,auto=None,use_best=False,early=200,depth=2,learn_rate=0.04,iterations=250):
    
    from catboost import CatBoostClassifier
    model = CatBoostClassifier(  
        iterations=iterations,    
        depth=depth,   
        learning_rate=learn_rate,       
        auto_class_weights=auto ,
        eval_metric = 'Accuracy',
        loss_function = 'Logloss',
        early_stopping_rounds=early,
        random_state=2021,
        )
    if use_best:
        model.fit(trainPool,
        # cat_features= cat_features,
        eval_set=testPool,
        #auto_class_weights = auto,
        use_best_model=True,
        #plot=True,
        verbose = 100
        )
    else:            
        model.fit(trainPool,
        #auto_class_weights=auto,
        #eval_set=(X_test, y_test),
        #use_best_model=True,
        #plot=True,
        verbose = 300
        )
    
    return model
```

```python
all_features = ['häst','kusk','bana','kön','lopp_dist','dist','kr','bins','podds','vodds','pris','spår','start','streck','timestamp','senast','ålder','plac']
cat_features = ['häst','kusk','bana','kön']
sel_feat = ['lopp_dist','dist','kr','bins','podds','vodds','timestamp','senast','pris','spår','start','streck','ålder']
add_features = ['lopp_dist','kr','timestamp','senast','pris','spår','start','streck','ålder']
sel_feat = cat_features + add_features
```

```python
def remove_worst(worst_f,cat,add):
    wf=set(worst_f)
    
    while len(add)>0:
        if len(wf)==0:
            add.remove(add[0])
            return cat, add
            
        worst=max(wf, key=worst_f.count)
    
        if worst=='kön':
            print('remove kön från cat')
            cat.remove(worst)
            return cat,add
        if worst in cat:
            print('remove',worst,'från worst_f')
            while worst in worst_f:
                worst_f.remove(worst)
            wf=set(worst_f)
            continue

        print('remove',worst,'från add')
        add.remove(worst)   
        return cat, add 
```

```python
import time
cat_features = ['häst','kusk','bana','kön']
add_features = ['lopp_dist','dist','kr','timestamp','senast','pris','spår','start','streck','ålder']

from catboost import Pool
n=3 # antal veckor för varje test 
t0=time.time()
while len(add_features)>0:
   
    sel_feat = cat_features + add_features
    test_res = 'test_time'+str(30-len(sel_feat))+'.csv'

    train, test_all = initsplit_my_data(df,mitt_datum)
    X_test=test_all.loc[0:0]   # låna kolumnerna
    _,X_test = move_rows_from_A_to_B(test_all,X_test,n)

    worst_features=[]
    best_iteration=[]
    best_score=[]
    i=0
    t1 = time.time()
    while X_test.shape[0] > 0:
        t2=time.time()
        y_test = X_test.plac
        X_test.drop('plac',axis=1,inplace=True)

        y_train = train.plac
        X_train = train.drop('plac',axis=1)
        print(f'X_train {X_train.shape}, test_all {test_all.shape}, df {df.shape}')
        trainPool=Pool(X_train[sel_feat],y_train,cat_features=cat_features)
        testPool =Pool(X_test[sel_feat],y_test,cat_features=cat_features)
            
        model = catfit(trainPool,testPool,auto='Balanced',use_best=True,early=200,iterations=2000,depth=None,learn_rate=None)
        worst_features+=[model.get_feature_importance(prettified=True).iloc[-1,0]]

        best_iteration += [model.get_best_iteration()]
        best_score += [model.get_best_score()['validation']['Accuracy']]
        
        print(f'\niter {i+1} avg_score: {np.mean(best_score)}, freq. worst: {max(set(worst_features), key=worst_features.count)}')
        print(model.get_feature_importance(prettified=True).iloc[:3],'\nworst\n',model.get_feature_importance(prettified=True).iloc[-2:]) 

        # move one datum to X_train
        test_all,train = move_rows_from_A_to_B(test_all, train,1)
        
        X_test=test_all.loc[0:0]   # låna kolumnerna
        _,X_test = move_rows_from_A_to_B(test_all,X_test,n)
    
        print(f'\nX_test {X_test.shape} test_all {test_all.shape} train {train.shape} {test_all.shape[0]+train.shape[0]==len(df)} ')
        print(round(time.time()-t2,2),'sek')
        i+=1
        
    print(f'\nFinal avg_score: {sum(best_score)/len(best_score)}, freq. worst: {max(set(worst_feature), key=worst_feature.count)}')
    print(model.get_feature_importance(prettified=True))
    bs=pd.DataFrame({'bi': best_iteration,'bs': best_score})
    bs.to_csv(test_res, index=False)
    cat_features,add_features=remove_worst(worst_features,cat_features,add_features)
                        
    print(round(time.time()-t1,2),'sec  ENDED',test_res)

    break

print('ENDED PGM totalt',round(time.time()-t0,2),'sek')
    
```

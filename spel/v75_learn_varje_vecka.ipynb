```python
def kelly(X,y, proba, kassa, printa=False):
    # For simple bets with two outcomes, one involving losing the entire amount bet, 
    # and the other involving winning the bet amount multiplied by the payoff odds, the Kelly bet is:

    # f is the fraction of the current bankroll to wager; (i.e. how much to bet, expressed in fraction)
    #    hur mycket av hela kassan skalll satsas per häst. Om negativt så satsar man på icke vinst om det går
    # b is the net fractional odds received on the wager; (e.g. betting $10, on win, rewards $4 plus wager; then b=0.4)
    #    b är vodds-1
    # p is the probability of a win
    #    p är proba
    # 1-p is the probability of a loss

    # f = (proba*vodds - 1) / (vodds-1) 

    # Finally, we combine those three concepts together. First, filter out all the low-risk horses of the day, and calculate their return ratio. 
    # Based on simulated past investment results to set an optimal return threshold. 
    # When the return ratio is higher than the threshold, using the Kelly Criterion to determine what percentage of the fund should bet on.

    kelly = pd.DataFrame(columns=['vodds','plac','vann','proba','f','spela','insats','rätt','utdelning','vinst'])
    #kelly = pd.DataFrame(X[['häst''vodds']])
    kelly.vodds=X.vodds
    kelly['vann'] = y
    kelly['plac'] = X.plac
    kelly['proba'] = proba[:,1]
    kelly['f'] = (kelly.proba*kelly.vodds - 1) / (kelly.vodds-1)
    kelly['spela'] = kelly.f >0
    kelly['insats'] = kelly.spela * kelly.f * kassa
    kelly['rätt'] = kelly.spela & kelly.vann
    kelly['utdelning'] = (kelly.insats * kelly.vodds) * kelly.rätt
    kelly['vinst'] = kelly.utdelning - kelly.insats
    
    startade = X.shape[0]
    spelade = kelly.spela.sum()
    ant_rätt = kelly.rätt.sum()
    insats = kelly.insats.sum()
    utdelning = kelly.utdelning.sum()
    vinst = kelly.vinst.sum()
    vinstprocent = 100*vinst/insats
    rättprocent = 100*ant_rätt/spelade
    if printa:
        print('antal startade',X.shape[0])
        print('antal spelade',kelly.spela.sum(),'varav rätt',kelly.rätt.sum())
        print('insats',kelly.insats.sum())
        print('utdelning',kelly.utdelning.sum())
        print()
        print('vinst {0:1.2f}'.format(kelly.vinst.sum()))
        print('vinst% {0:1.2f} rätt% {1:1.2f}'.format(vinstprocent,rättprocent))

    return kelly, startade, spelade, ant_rätt, insats, utdelning, vinst, vinstprocent, rättprocent
```

```python
def själva_loopen(start_datum, modeller, selected_features,cat_features,iterations=2000):
    other_features = ['datum','avd','plac']
    # först selektera features
    #mer_komplett.plac = (mer_komplett.plac==1)*1
    mer_komplett.sort_values(by=['datum','bana','avd'],inplace=True)
    mer_komplett.reset_index(drop=True,inplace=True)
                        
    df = mer_komplett[selected_features + other_features].drop([],axis=1).copy(deep=True)
    df['vodds'] = mer_komplett.vodds.copy()
    # sätt kassan och initiera en resultat_df
    kassa=1000
    print('kassa',kassa)
    resultat_df = pd.DataFrame(columns=['datum','startade','spelade','rätt','insats','utdelning','vinst','vinstprocent','rättprocent'])  # per omgång
    kelly_df =    pd.DataFrame(columns=['datum','vann','plac','proba','f','spela','insats','rätt','utdelning','vinst'])   # allt per häst

    # skapa träningsdata med selected features och spara resten 
    ## One week in week for week training
    resten=df[df.datum > start_datum]
    
    train = df[df.datum <= start_datum].copy(deep=True)
    datum_rest=resten.iloc[0].datum if resten.shape[0] > 0 else None
    datum_train = start_datum
    y_train = train.plac
    #train.drop('plac',axis=1,inplace=True)
    y_test = resten.plac
    #resten.drop('plac',axis=1,inplace=True)
        
    resten.sort_values(by=['datum','bana'],inplace=True)
    train.sort_values(by=['datum','bana'],inplace=True)

    #train,resten=dela_på_mitten(mer_komplett[selected_features])
    print('train size =',train.shape[0],'resten size =',resten.shape[0])
    #print('kolla ok', 0==mer_komplett.shape[0]-train.shape[0]-resten.shape[0])
    n=0
    while resten.shape[0] > 0:
        n+=1 
        datum_rest=resten.iloc[0].datum
        datum_train = train.iloc[-1].datum

        print('train - sista datum',datum_train)
        print('resten - första datum', datum_rest)
    
        # Bestäm train och test med sin första datum 
        # X_train = train.drop(other_features,axis=1)
        X_train = train[selected_features]
        y_train = train.plac

        test = resten[resten.datum==datum_rest]

        # if 'vodds' not in test.columns.tolist():
        #     print(resten.info())

        # X_test = test.drop(other_features,axis=1)
        X_test=test[selected_features]
        y_test = test.plac
        # print('\ntrain info')
        # print(train.info())
        # print('\nresten info')
        # print(resten.info())
        # print('\ntest info')
        # print(test.info())
        # print('\nX_test info')
        # print(X_test.info())

        if mer_komplett.shape[0]-train.shape[0]-resten.shape[0]:
            print('\n\nSHAPE ERROR\n\n')
            break

        #fit
        y_train=(y_train==1)*1
        y_test=(y_test==1)*1
        model = catfit(cat_features, X_train, y_train, X_test,y_test,iterations=iterations)
  
        # predict
        predicted, proba, feat = pred(model,X_test,y_test)

        # kelly    
        print(f'df shape {df.shape} X_test.shape {X_test.shape}')
        X_tst=test.copy()
        X_tst['vodds'] = df.vodds
        kelly_temp, startade, spelade, ant_rätt, insats, utdelning, vinst, vinstprocent, rättprocent = kelly(X_tst,y_test,proba, 500)
        kelly_temp['datum']=datum_rest
        kelly_temp['kassa'] = kassa
        kelly_df = pd.concat([kelly_df,kelly_temp])

        resultat_df.loc[resultat_df.shape[0]] = datum_rest,startade,spelade,ant_rätt,insats,utdelning,vinst,vinstprocent,rättprocent

        #uppdatera kassan 
        print("Vinst {0:1.1f} vinst% {1:1.1f} spelade {2:d} av {3:d} startade varav {4:d} rätt, rätt% {5:1.1f}".format(vinst, vinstprocent,spelade,startade,ant_rätt, rättprocent))    

        kassa = kassa+vinst
        print(f'\n{n} kassa: {round(kassa,1)}')

        spara_modell(model, modeller+'/model_'+str(datum_train)[:10])        
        print()
        # flytta senaste spel till train
        train, resten = flytta_ett_datum(train,resten,datum_rest)
    

    print('Kassa vid spelets slut', kassa)

    datum_train = train.iloc[-1].datum
        
    kelly_df.to_csv('C:\\Users\\peter\Documents\\MyProjects\\PyProj\\Trav\\kelly_df_1000.csv', index=False)
    resultat_df.to_csv('C:\\Users\\peter\Documents\\MyProjects\\PyProj\\Trav\\resultat_df_1000.csv', index=False)

    # Bestäm train och test med sin första datum 
    # X_train = train.drop(other_features,axis=1)
    X_train = train[selected_features]
    y_train = train.plac
    # X_test = resten[resten.datum==datum_rest].drop(other_features,axis=1)
    X_test = resten[resten.datum==datum_rest][selected_features]
    y_test = resten[resten.datum==datum_rest].plac
    print('avslutande train shape',train.shape)
    print('avslutande resten shape',resten.shape)
    print('avslutande X_test shape',X_test.shape)
    y_train=(y_train==1)*1
    model = catfit(cat_features, X_train, y_train, X_test,y_test,iterations=iterations)

    print('spara model senaste modell')
    spara_modell(model, modeller+'/model_'+str(datum_train)[:10])        
    spara_modell(model, modeller+'/model_senaste')

    return model
```

```python
def kör(modeller, iterations, selected_features, cat_features,start_datum):

    print('starta', modeller, 'från datum', start_datum)
    print(selected_features,cat_features)
    model = själva_loopen(start_datum,modeller,selected_features,cat_features, iterations=iterations)

    model.get_feature_importance(prettified=True)
```
